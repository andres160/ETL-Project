```python
from bs4 import BeautifulSoup
from splinter import Browser
import os
import pandas as pd
import time
from pymongo import MongoClient
client = MongoClient()
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

url = "https://rateyourmusic.com/customchart?page=1&chart_type=top&type=single&year=alltime&genre_include=1&include_child_genres=1&genres=Salsa&include_child_genres_chk=1&include=both&origin_countries=&limit=none&countries="
browser.visit(url)
```

```python
title_list = []
artist_list = []

def get_data():
    for x in range(1,5):
        # HTML object
        html = browser.html
        # Parse HTML with Beautiful Soup
        soup = BeautifulSoup(html, 'html.parser')
        articles = soup.find_all(class_='chart_main')

        for article in articles:        
            artist = article.find(class_='artist')
            title = article.find(class_='album')
            print('-----------')
            title_list.append(title.string)      
            artist_list.append(artist.string) 
            
        
        browser.click_link_by_partial_text('next >')
        time.sleep(5)
```

```python
ratings_list = []

def get_data1():
    for x in range(1,5):
        # HTML object
        html = browser.html
        # Parse HTML with Beautiful Soup
        soup = BeautifulSoup(html, 'html.parser')
        ratings = soup.find_all(class_='chart_stats')

        for rating in ratings:        
            number = rating.find("b")
            print('-----------')
            ratings_list.append(number.string)             
        
        browser.click_link_by_partial_text('next >')
        time.sleep(5)
```
